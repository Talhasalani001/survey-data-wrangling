```python
def clean_data(new_df):
    new_df['Q1.002'] = new_df['Q1.002'].fillna(0)
    new_df['Q1.004'] = new_df['Q1.004'].fillna(0)
    
    new_df['Access Code'] = new_df['Access Code'].fillna('Unknown')
    new_df['Email Address'] = new_df['Email Address'].fillna('Unknown')
    
    new_df['Region'] = new_df['Region'].fillna('Unknown')
    new_df['City / Town / District'] = new_df['City / Town / District'].fillna('Unknown')
    
    gender_columns = ['Q2.Male', 'Q2.Female', 'Q2.NonBinary', 'Q2.PreferNotToAnswer']
    new_df[gender_columns] = new_df[gender_columns].fillna(0)

    new_df['Q3.Age'] = pd.to_numeric(new_df['Q3.Age'].str.replace(' ', ''), errors='coerce')

   

    new_df.loc[27, 'Q4.Height'] = 61
    new_df.loc[32, 'Q4.Height'] = 70

    new_df['Q5.CountryOfCitizenship'] = new_df['Q5.CountryOfCitizenship'].str.title()


    redundancy_map = {
    'Us': 'United States',
    'Usa': 'United States',
    'United States ': 'United States', 
    'United States ; Pakistan ': 'United States'  
    }

    new_df['Q5.CountryOfCitizenship'] = new_df['Q5.CountryOfCitizenship'].replace(redundancy_map)

    new_df['Q6.UndergradDegree'] = new_df['Q6.UndergradDegree'].str.title()


    personal_columns = ['Q4.Height', 'Q5.CountryOfCitizenship', 'Q6.UndergradDegree', 'Q7.ExpectedGraduation']
    new_df[personal_columns] = new_df[personal_columns].fillna('Not Provided')
    
    device_columns = ['Q8.MicrosoftWindows', 'Q8.AppleMacBook', 'Q8.Other']
    new_df[device_columns] = new_df[device_columns].fillna(0)

    
    
    if 'Unnamed: 26' in df.columns:
        new_df.drop(columns=['Unnamed: 26'], inplace=True)
    
    survey_columns = ['Q9.YesFullTime', 'Q9.WorkingButNotFullTime', 'Q9.NotWorking', 'Q10.Little/None', 
                      'Q10.SomeFamiliarity', 'Q10.AverageUser', 'Q10.FrequentUseForproj', 'Q10.FluentExpert', 
                      'Q11.Little/none', 'Q11.Some familiarity', 'Q11.Average user', 'Q11.Frequent use for projects', 
                      'Q11.Fluent/expert', 'Q12.Little/none', 'Q12.Some familiarity', 'Q12.Average user', 
                      'Q12.Frequent use for projects', 'Q12.Fluent/expert', 'Q13.Yes', 'Q13.No', 
                      'Q14.Yes', 'Q14.No', 'Q14.NotSure']
    new_df[survey_columns] = new_df[survey_columns].fillna(0)
    
    new_df['Q15.GoalForLearningDataAnalytics'] = new_df['Q15.GoalForLearningDataAnalytics'].fillna('Not Provided')
    
    return new_df

cleaned_df = clean_data(new_df)
```

### Discussion: 

#### To handle missing values in the dataset, each column was examined based on its content and purpose. For identifier and contact details such as 'Access Code' and 'Email Address', missing values were filled with 'Unknown' to avoid making incorrect assumptions. Geographic information like 'Region' and 'City / Town / District' was also filled with 'Unknown' due to the wide variability and lack of a logical mean or mode for imputation. Columns with categorical data, like 'Q5.CountryOfCitizenship', were filled with 'Not Provided' to maintain data integrity without introducing biases. In numeric survey response columns (e.g., Q8, Q9), missing values were replaced with 0, reflecting non-participation or zero count, which aligns with the survey's intent.

#### General data cleaning involved removing white spaces, standardizing casing, and consolidating redundant entries to ensure consistency and readability. This approach prioritizes transparency, avoiding assumptions that could skew the data. By using placeholders like 'Unknown' or 'Not Provided', the dataset maintains clarity about missing data while staying prepared for further analysis or modeling. Numeric fills such as 0 were used where appropriate, preserving the logical structure of the data without altering its distribution significantly.
